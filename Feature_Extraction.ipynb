# Axial Scan Radiomic Feature Extraction

## Imports

```python
import numpy as np
import pandas as pd
import six
import os

from radiomics import featureextractor
from extraction import get_features
from extraction import mask_path
```

## Specify Data and Mask Directories

```python
cwd = os.getcwd()
cwd
```

```python
data_dir = cwd + '/Axial Images/'
```

```python
images = os.listdir(data_dir)
if '.DS_Store' in images:
    images.remove('.DS_Store')
```

## Feature Extraction

```python
extractor = featureextractor.RadiomicsFeatureExtractor()
```

```python
result = get_features(extractor, data_dir + images[0], mask_path(images[0], 0))
features = list(result.keys())[22:] # '0:21' diagnostic features, '22:' radiomic features
```

```python
dfs = {}
rois = [0, 1, 2, 6, 10, 12, 13]
```

```python
for roi in rois:
    feature_space = []
    for img in images:
        row = [img]
        imagePath = data_dir + img
        maskPath = mask_path(img, roi)
        #print('\nProcessing...')
        #print('Image Path:', imagePath)
        #print('Mask Path:', maskPath) 
        result = get_features(extractor, imagePath, maskPath)
        for i in range(len(features)):
            row.append(result[features[i]])
        feature_space.append(row)
    df = pd.DataFrame(feature_space, columns = ["Image"] + features).set_index("Image")
    dfs[roi] = df
```

```python
dfs[0].to_csv('Axial_313_Features.csv')
dfs[1].to_csv('Axial_313-1_Features.csv')
dfs[2].to_csv('Axial_313-2_Features.csv')
dfs[6].to_csv('Axial_313-6_Features.csv')
dfs[10].to_csv('Axial_313-10_Features.csv')
dfs[12].to_csv('Axial_313-12_Features.csv')
dfs[13].to_csv('Axial_313-13_Features.csv')
```
